```python
1. Create a line chart to visualize the distribution of the 'capitalLong' feature
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('spambase(2).csv')

# Plotting the Kernel Density Estimate (KDE) for 'capitalLong' to show its distribution
plt.figure(figsize=(10, 6))

# Create the KDE plot (smoothed distribution)
sns.kdeplot(df['capitalLong'], fill=True, color='blue', linewidth=2)

# Customizing the plot
plt.title('Distribution of Capital Long Feature (KDE)')
plt.xlabel('Capital Long')
plt.ylabel('Density')
plt.grid(True)

# Show the plot
plt.show()


```

```python
6. Generate a table chart to display the summary statistics for the 'capitalAve' feature.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('spambase(2).csv')

summary_stats = df['capitalAve'].describe()

summary_df = pd.DataFrame(summary_stats).transpose()

fig, ax = plt.subplots(figsize=(8, 4))  # Size of the table

ax.axis('off')

table = ax.table(cellText=summary_df.values, colLabels=summary_df.columns, loc='center', cellLoc='center', bbox=[0, 0, 1, 1])

table.auto_set_font_size(False)
table.set_fontsize(12)
table.auto_set_column_width(col=list(range(len(summary_df.columns))))  # Adjust column width based on content

plt.title('Summary Statistics for CapitalAve')
plt.show()
```

```python
18. Visualize the correlation matrix using a heatmap.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('spambase(2).csv')

# Ensure we are only using numeric columns for correlation
numeric_columns = df.select_dtypes(include=['number']).columns

# Calculate the correlation matrix for all numeric features
correlation_matrix = df[numeric_columns].corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))  # Adjust the size of the plot

# Create a heatmap with annotation (show the correlation values in the cells)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True, 
            linewidths=0.5, square=True, vmin=-1, vmax=1)

# Add title
plt.title('Correlation Matrix Heatmap')

# Display the plot
plt.show()
```

```python
23. Describe the characteristics of data that make a bar chart suitable for visualization
Bar charts are useful when we want to compare different groups or categories of data, making them good for categorical data such as 
gender or hair color.
```

```python
24. Explain when it is appropriate to use a scatter plot for data exploration.
Scatter plots are appropriate to show the relationship between two variables to display if there is a positive or negative 
correlation between them.
```

```python
25. Discuss the advantages and disadvantages of using a pie chart for categorical data
Advantages- Pies charts are visually appealing and easy to understand data just by looking how data fits into the pie. We can make comparisons
about data by looking how big each "slice" is.
Disadvantages- Pie charts are best when there are not many categories, preference for 5-7 different categories and cannot display large
    amounts of data. The "slices" in the pie may be visually similar, so it can difficult to decipher how the data actually stacks up
    against each other.
```

```python
26. When would you choose a line chart over a histogram for data visualization?
Line charts are used to track change over time and how they fluctuate. We may want to use a line chart when talking about daily 
temperature changes.
```

```python
27. Describe a scenario where a table chart is more informative than other chart types.
Table charts can display large amounts of data. A scenario where this could be useful is for budget planning. A family may want
to track their monthly expenses and they list different categories such as food, housing, recreation, clothing, and transportation.
```

```python
28. Explain the key insights you can gain from a histogram.
Histograms show distribution of data and display the mean, median and mode. We can see if the data is skewed (asymmetric) or if
there are any outliers that do not fit the data trend.
```
